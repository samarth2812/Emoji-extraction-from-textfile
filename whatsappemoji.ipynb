```python
import emoji
import pandas as pd
from collections import Counter
import matplotlib.pyplot as plt
```

Extracting emojis from the file

```python
def extract_emojis(text):
    emojis = []
    for character in text:
        if character in emoji.UNICODE_EMOJI["en"]:
            emojis.append(character)
    return emojis
```

```python
def get_emoji_shortcodes(emojis):
    emoji_shortcodes = []
    for emoji_symbol in emojis:
        shortcode = emoji.demojize(emoji_symbol)
        emoji_shortcodes.append(shortcode)
    return emoji_shortcodes
```

```python

file_path = 'r.txt'
with open(file_path, 'r', encoding='utf-8') as file:
    content = file.read()
    emojis = extract_emojis(content)
    emoji_shortcodes = get_emoji_shortcodes(emojis)
    print(emoji_shortcodes)
    # Print the shortcodes
    # for shortcode in emoji_shortcodes:
    #     print(shortcode)
```

```python
label_tag=[]
def label_emojis(emoji_list):
    emoji_sentiment = {
        'ü•≤':'Positive',
        'üòä': 'Positive',
        'üòÑ':'Positive',
        'üòÉ':'Positive',
        'üåü':'Positive',
        'üéâ':'Positive',
        '‚ù§Ô∏è':'Positive' ,
         'üòç':'Positive',
         'üòÅ':'Positive',
          'üíñ':'Positive',
          'üòÇ':'Positive',
          'ü§£':'Positive',
        'üò¢': 'Negative',
        'üò≥':'Neutral',
        'üëç': 'Positive',
        'üëé': 'Negative',
        'üòê': 'Neutral',
        # Add more emojis and their corresponding labels here
    }
    labeled_emojis = []
    for emoji in emoji_list:
        if emoji in emoji_sentiment:
            label = emoji_sentiment[emoji]
        else:
            label = 'Unknown'
        labeled_emojis.append((emoji, label))

    return labeled_emojis
labeled_emojis = label_emojis(emojis)
for emoji, label in labeled_emojis:
    label_tag.append(label)
print(label_tag)
```
